```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("the food is wow")
```

```python
classifier(["the food is wow","this is very bad"])
```

```python
classifier = pipeline("zero-shot-classification")
classifier("the course is very usefull", candidate_labels=['education','business','shopping'])
```

```python
generator = pipeline("text-generation")
generator("to be or not to be")
```

```python
generator = pipeline("text-generation", model='distilbert/distilgpt2')
generator("to be or not to be", max_length=30,num_return_sequences=2)
```

```python
unmasker = pipeline("fill-mask")
unmasker("the sky today was <mask>", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Ahmad and I live in Jeddah")
```
